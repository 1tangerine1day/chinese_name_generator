```python
def get_char():
    list_char = list()
    for i in range(0x4e00,0x9fa6):
        list_char.append(chr(i))
    return list_char
```

```python
list_char = get_char()
```

```python
def check_is_ch_word(word) : 
    for w in word :
        if w not in list_char :
            return False
        
    return True
```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNN, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        self.gru = nn.GRU(self.input_size, self.hidden_size)
        self.h2o = nn.Linear(self.hidden_size, self.output_size)
        self.softmax = nn.LogSoftmax()
        
    def forward(self, input):
        out,_ = self.gru(input)
        out = self.h2o(out).squeeze(0).squeeze(0)
        output = self.softmax(out)
        return output

    def initHidden(self):
        return torch.zeros(1, self.hidden_size)
```

```python
def predict(char):
    test = dict_char[char]
    test_tensor = torch.from_numpy(test).unsqueeze(0).unsqueeze(0)
    #print (test_tensor.shape)
    gen_test = model(test_tensor)
    torch.max(gen_test,0)[1]
    
    k = int(torch.max(gen_test,0)[1])
    
    return dict_idx_char[k]
```

```python
def predict_seq(char) :
    name = char
    count = 1
    
    char = predict(char)
    name += char
    count += 1
    
    while can_stop(char, count) is False :
        char = predict(char)
        name += char
        count += 1
    return name
```

```python
def can_stop(char, count):
    re = True
    if char == '。': 
        re = False
        print("ok1")
    if count < 10 :
        re = False
        print("ok2")
    return re
```
